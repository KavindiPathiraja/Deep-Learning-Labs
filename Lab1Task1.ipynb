```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from google.colab import drive

drive.mount('/content/drive')

pokemon_path = '/content/drive/MyDrive/MyDrive/DL/Lab1/Pokemon.csv'
```

```python
exp_array = np.random.exponential(scale=1.0, size=(4,4))
print("4x4 Exponential Array:\n", exp_array)
```

```python
exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.rand(100000)
norm_data = np.random.randn(100000)

plt.figure(figsize=(8,5))
plt.hist(exp_data, density=True, bins=200, histtype="step", color="red", label="Exponential")
plt.hist(uni_data, density=True, bins=200, histtype="step", color="blue", label="Uniform")
plt.hist(norm_data, density=True, bins=200, histtype="step", color="green", label="Normal")
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

```python
df_pokemon = pd.read_csv(pokemon_path)

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_corr_pearson = df_pokemon[features].corr(method='pearson')
df_corr_spearman = df_pokemon[features].corr(method='spearman')

plt.figure(figsize=(8,6))
sns.heatmap(df_corr_pearson, annot=True, cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df_corr_spearman, annot=True, cmap='coolwarm')
plt.title("Spearman Correlation Heatmap")
plt.show()
```
